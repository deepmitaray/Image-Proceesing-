```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread(r'C:\Users\DEEPMITA RAY\Pictures\Screenshots\ironman.jpeg')
```

```python
c = 255 / np.log(1 + np.max(image))
log_image = c * (np.log(image + 1))
```

```python
log_image = np.array(log_image, dtype = np.uint8)
```

```python
# Display both images
plt.imshow(image)
plt.show()
plt.imshow(log_image)
plt.show()
```
